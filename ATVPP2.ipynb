```python
medias_categorias= data.groupby(by = 'CATEGORIAS') ['MEDV'].mean()
medias_categorias
```

```python
dic_baseline = {
    'Grande': medias_categorias[0],
    'Médio': medias_categorias[1],
    'Pequeno': medias_categorias[2]
}
```

```python
dic_baseline
```

```python
def retorna_baseline(num_quartos):
    if num_quartos <= 4:
        return dic_baseline.get('Pequeno')
    elif num_quartos < 7:
        return dic_baseline.get('Médio')
    else:
        return dic_baseline.get('Grande')
```

```python
data.columns
```

```python
y = data['MEDV']
x = data.drop(['TOWN', 'TRACT', 'LON', 'LAT', 'MEDV', 'ZN', 'DIS', 'TAX', 'CATEGORIAS'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state=42)
```

```python
y_pred = []
for i in x_test.RM.items():
  n_quartos = i[1]
  y_pred.append(retorna_baseline(n_quartos))
```

```python
df_resultados = pd.DataFrame()
df_resultados['y_teste_real'] = y_test.values
df_resultados['y_pred_baseline'] = y_pred
df_resultados.head()
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_resultados.index,
                         y=df_resultados.y_teste_real,
                         mode='lines+markers',
                         name='valor real'))
fig.add_trace(go.Scatter(x=df_resultados.index,
                         y=df_resultados.y_pred_baseline,
                         mode='lines+markers',
                         name='valor predito baseline'))
fig.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_log_error
from sklearn.metrics import root_mean_squared_error

def valida_modelo(nome, y_pred, t_test):
  print('Validação do modelo:', nome)
  r2 = r2_score(y_pred, t_test)
  print('R2:', r2)
  mae = mean_absolute_error(y_pred, t_test)
  print('MAE:', mae)
  mse = mean_absolute_error(y_pred, t_test)
  print('MSE:', mse)
  rmse = np.sqrt(mse)
  print('RMSE:', rmse)
  mape = mean_absolute_percentage_error(y_pred, t_test)
  print('MAPE:', mape)
```

```python
valida_modelo('baseline', y_test, y_pred)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)
```

```python
valida_modelo('Regressão Linear', lr_pred, y_test)
```

```python
df_resultados['y_pred_lr'] = lr_pred
df_resultados.head()
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_resultados.index,
                         y=df_resultados.y_teste_real,
                         mode='lines+markers',
                         name='valor real'))
fig.add_trace(go.Scatter(x=df_resultados.index,
                         y=df_resultados.y_pred_baseline,
                         mode='lines+markers',
                         name='valor predito baseline'))
fig.add_trace(go.Scatter(x=df_resultados.index,
                         y=df_resultados.y_pred_lr,
                         mode='lines+markers',
                         name='valor predito baseline'))
fig.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
dt_pred = dt.predict(x_test)
```

```python
valida_modelo('arvore de decisão', dt_pred, y_test)
```

```python
df_resultados['y_pred_dt'] = dt_pred
df_resultados.head()
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_resultados.index,
                         y=df_resultados.y_teste_real,
                         mode='lines+markers',
                         name='valor real'))
fig.add_trace(go.Scatter(x=df_resultados.index,
                         y=df_resultados.y_pred_baseline,
                         mode='lines+markers',
                         name='valor predito baseline'))
fig.add_trace(go.Scatter(x=df_resultados.index,
                         y=df_resultados.y_pred_lr,
                         mode='lines+markers',
                         name='valor predito regressão linar'))
fig.add_trace(go.Scatter(x=df_resultados.index,
                         y=df_resultados.y_pred_dt,
                         mode='lines+markers',
                         name='valor predito dt'))
fig.show()
```
